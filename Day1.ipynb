```python
import pandas as pd
import matplotlib.pyplot as plt



today_consumption = df["Today's Consumption(KW.h)"].tolist()
total_consumption = df["In Total(KW.h)"].tolist()


ind = range(len(today_consumption))


width = 0.35
plt.figure(figsize=(10, 6))

plt.bar(ind, today_consumption, width, label="Today's Consumption", color='blue', edgecolor='black')
plt.bar(ind, total_consumption, width, label='Total Consumption', color='green', edgecolor='black', alpha=0.5)

plt.xlabel('Sample')
plt.ylabel('KW.h')
plt.title("Today's and Total Consumption (KW.h)")


plt.legend()
plt.show()
```
